```python
filepath = ('data/temperature_2014.csv')
df4 = pd.read_csv(filepath)
# df4.info()
df4['PST'] = pd.to_datetime(df4['PST'])
# df4.info()
fig = plt.figure(figsize=(10, 6))
plt.plot(df4['PST'], df4['Max TemperatureF'], c='red', alpha=0.5)
plt.plot(df4['PST'], df4['Min TemperatureF'], c='green', alpha=0.5)
plt.title('온도통계', fontsize=20)
plt.xlabel('날짜', fontsize=16)
plt.ylabel('기온', fontsize=16)
plt.tick_params(axis='both', labelsize=12)
plt.fill_between(df4['PST'], df4['Max TemperatureF'], df4['Min TemperatureF'], facecolor='green', alpha=0.2)
plt.savefig('weatherFig.png')
plt.show()
```
