```python
from nltk import Text
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

path      = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12, 9)
```

```python
kolawtext = Text(okt.nouns(c))
kolawtext.plot(30)
```

```python
kv = kolawtext.vocab()
kv
```

```python
kv2 = dict()

for tag, count in kv.items():
    if len(tag) > 1:
        kv2[tag] = count

kv2
```

```python
from wordcloud import WordCloud

path      = 'c:/Windows/Fonts/malgun.ttf'
wc = WordCloud(width = 1000, height = 800, background_color = 'white', random_state = 0, font_path = path)
plt.axis('off')
plt.imshow(wc.generate_from_frequencies(kv2))
```

```python
from collections import Counter

s2 = "aaaabbcc"
sc2 = Counter(s2)

sc2
```

```python
nouns = okt.nouns(c)
nounscounter = Counter(nouns)
print(nounscounter)
print(nounscounter.most_common(10))
```

```python
lawcounter = dict()

for tag, count in nounscounter.items():
    if len(tag) > 1 and count > 1:
        lawcounter[tag] = count

print(lawcounter)
```

```python
wc2 = WordCloud(width=1000, height=800, background_color='white', random_state=0, font_path=path)
plt.axis('off')
plt.imshow(wc2.generate_from_frequencies(lawcounter))
```

```python
news = open('data/newsdata.txt').read()
news_nouns = okt.nouns(news)
news_nouns_counter = Counter(news_nouns)
news_dict = dict()

for k, v in news_nouns_counter.items():
    if len(k) > 1 and v > 1:
        news_dict[k] = v

print(news_dict)

wc2 = WordCloud(width=500, height=400, background_color='white', random_state=0, font_path=path)
plt.axis('off')
plt.imshow(wc2.generate_from_frequencies(news_dict))
```
