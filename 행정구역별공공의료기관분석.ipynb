```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/보건복지부_공공보건 의료기관 현황.csv', encoding='CP949')
addr = pd.DataFrame(data['주소'].apply(lambda x : x.split()[:2]).to_list(), columns=['시도', '군구'])

addr.head()
```

```python
addr.loc[addr['시도'] == '창원시']
addr.iloc[27] = ['경상남도', '창원시']
addr.iloc[31] = ['경상남도', '창원시']

addr['시도'].value_counts()
```

```python
addr.loc[addr['시도'] == '경산시']
addr.iloc[47] = ['경상북도', '경산시']

addr['시도'].value_counts()
```

```python
addr.loc[addr['시도'] == '천안시']
addr.iloc[209] = ['충청남도', '천안시']
addr.iloc[210] = ['충청남도', '천안시']
```

```python
goongoo_alias = {'경남' : '경상남도', '서울시' : '서울특별시', '경북' : '경상북도', '충북' : '충청북도', '충남' : '충청남도', 
                 '전남' : '전라남도', '경기' : '경기도', '전북' : '전라북도', '대전시' : '대전광역시', '부산특별시' : '부산광역시',
                 '강원도' : '강원특별자치도'}

addr['시도'] = addr['시도'].apply(lambda x : goongoo_alias.get(x, x))

addr['시도'].value_counts()
```

```python
addr.loc[addr['군구'] == '아란13길']
addr.iloc[75] = ['제주특별자치도', '제주시']
addr.loc[addr['군구'] == '동매로']
addr.iloc[155] = ['부산광역시', '사하구']

addr['군구'].unique()
```

```python
addr['시도군구'] = addr.apply(lambda x : x['시도'] + " " + x['군구'], axis = 1)
addr['count'] = 0
addr_groupby = pd.DataFrame(addr.groupby(by=['시도', '군구', '시도군구'], as_index=False).count())
addr_groupby = addr_groupby.set_index('시도군구')
addr_groupby
```

```python
population = pd.read_csv('data/행정구역_시군구_별__성별_인구수.csv', encoding='CP949')
population['군구'] = population['군구'].str.strip()
population['군구'].unique()

population['시도군구'] = population.apply(lambda x : x['시도'] + " " + x['군구'], axis = 1)
population = population.loc[population['시도'] != population['군구']].reset_index(drop=True).set_index('시도군구')
population
```

```python
addr_population = pd.merge(addr_groupby, population[['총인구수', '남자인구수', '여자인구수']], how='inner', on='시도군구')
addr_population
```

```python
data_draw_kor = pd.read_csv('data/data_draw_korea.csv')
data_draw_kor = data_draw_kor.iloc[::, 1::]

data_alias = {'강원도' : '강원특별자치도'}

data_draw_kor['광역시도'] = data_draw_kor['광역시도'].apply(lambda x : data_alias.get(x, x))
data_draw_kor['광역시도'].unique()
```
