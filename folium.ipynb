```python
lat1 = 37.5667
lon1 = 126.78006

lat2 = 37.5658859
lon2 = 126.9754788

m = folium.Map(location=[lat1, lon1], zoom_start=10)

folium.Marker([lat1, lon1], popup='김포공항', icon=folium.Icon(color='green', icon='plane', icon_color='yellow')).add_to(m)
folium.Marker([lat2, lon2], popup='덕수궁', icon=folium.Icon(color='red', icon='cloud', icon_color='skyblue')).add_to(m)
m
```

```python
location00 = [35.1795543, 129.0756416]
location10 = [35.1805543, 129.0756416]
location11 = [35.1785543, 129.0756416]
location12 = [35.1795543, 129.0746416]
location13 = [35.1795543, 129.0766416]
lines = [location00, location10, location11, location12, location13]
m2 = folium.Map(location=location00, zoom_start=15)
folium.PolyLine(lines, tooltip='Coast').add_to(m2)
folium.Marker(location=location00, popup='부산시청').add_to(m2)
folium.Marker(location=location10, popup='위').add_to(m2)
folium.Marker(location=location11, popup='아래').add_to(m2)
folium.Marker(location=location12, popup='왼쪽').add_to(m2)
folium.Marker(location=location13, popup='오른쪽').add_to(m2)

m2
```
