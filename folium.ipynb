```python
# 좌표 100개 생성
import numpy as np
from folium import plugins
n=100
data=np.array([np.random.uniform(low=35,high=38, size=n),np.random.uniform(low=126,high=129, size=n)]).T
data
# pop은 문자열로 저장되어짐
popup=[str(i) for i in range(1, n+1)]
# 지도 줌레벨에 따라서  합산수 산출 최대시 개별
m=folium.Map(location=[lat11,lon11],zoom_start=7)
# 플러그인 사용해서 데이터좌표 수
plugins.MarkerCluster(data, popups=popup).add_to(m)
m
```

```python
import pandas as pd
df=pd.read_csv('data/CCTV_20190917.csv')
df
```

```python
# 데이터 정제(축소)
df1=df.loc[df['카메라대수']>0,['소재지도로명주소','위도','경도']]
df1
```

```python
# 팝업 리스트
# 시나리오
popups=df1['소재지도로명주소'].to_list()
popups
# 위도경도 리스트화
lat=df1['위도'].to_list()
# lat
lng=df1['경도'].to_list()
# lng

data=np.array([lat,lng]).T
# 평균 좌표를 기준점으로
lat_avg=np.array(lat).mean()
lng_avg=np.array(lng).mean()
print(lat_avg,lng_avg)

m=folium.Map(location=[lat_avg,lng_avg],zoom_start=14)
plugins.MarkerCluster(data,popups=popups).add_to(m)
m
```

```python
df2 = df.loc[(df['설치목적구분'] == '어린이보호') & (df['카메라대수'] > 0)]

df2_lat   = df2['위도'].to_list()
df2_lon   = df2['경도'].to_list()
df2_popup = df2['소재지도로명주소'].to_list()
#df2_lat
#df2_lon

locations = np.array([df2_lat, df2_lon]).T
#locations

df2_lat_avg = locations[:, 0].mean()
df2_lon_avg = locations[:, 1].mean()
#df2_lat_avg
#df2_lon_avg

mymap = folium.Map(location=[df2_lat_avg, df2_lon_avg], zoom_start=14)
plugins.MarkerCluster(locations, popups=df2_popup).add_to(mymap)
mymap
```
