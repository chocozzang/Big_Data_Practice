```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager

sns.set(style = 'whitegrid')
path      = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12, 9)
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'smoker')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin', hue = 'smoker', split = True, inner = 'stick')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'boxen')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', kind = 'bar')
```
