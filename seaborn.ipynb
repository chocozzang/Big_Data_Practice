```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager

sns.set(style = 'whitegrid')
path      = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12, 9)
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
tips    = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(y = 'class', data = titanic, hue="who", palette="YlGn")
```

```python
# sns.countplot(x='day', data=tips, hue='sex')
# sns.countplot(x='day', data=tips, hue='smoker', palette="YlGn")
sns.countplot(x='day', data=tips, hue='time', palette='Accent_r')
```

```python
# color_palette = sns.color_palette("terrain_r", 10)
color_palette = sns.color_palette("CMRmap", 10)
sns.palplot(color_palette)
plt.show()
```

```python
print(tips['total_bill'].corr(tips['tip']))
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue='sex')
plt.title("tips")
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue='smoker', style='sex')
plt.title("tips")
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue='size', style='size')
plt.title("tips")
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue='sex', col='day', col_wrap=2)
plt.title("tips")
plt.show()
```
